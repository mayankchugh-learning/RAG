```python
import requests
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
```

```python
url="https://raw.githubusercontent.com/langchain-ai/langchain/master/docs/docs/modules/state_of_the_union.txt"
```

```python
res = requests.get(url)
```

```python
with open("state_of_the_union.txt", "w") as f:
    f.write(res.text)
```

```python
loader=TextLoader("./state_of_the_union.txt")
```

```python
documents=loader.load()
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter=CharacterTextSplitter(chunk_size=500,chunk_overlap=50)
```

```python
chunks=text_splitter.split_documents(documents)
```

```python
chunks[2].page_content
```

```python
len(chunks)
```

```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Weaviate
import weaviate
from weaviate.embedded import EmbeddedOptions
```

```python
from google.colab import userdata
OPENAI_API_KEY=userdata.get('OPENAI_API_KEY')
```

```python
import os
os.environ["OPENAI_API_KEY"]=OPENAI_API_KEY
```

```python
client=weaviate.Client(
    embedded_options=EmbeddedOptions()
)
```

```python
vectorstore = Weaviate.from_documents(
    client=client,
    documents=chunks,
    embedding=OpenAIEmbeddings(),
    by_text=False
)
```

```python
retriever = vectorstore.as_retriever()
```

```python
from langchain.chat_models import ChatOpenAI
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
llm=ChatOpenAI(model_name="gpt-3.5-turbo",temperature=0.2)
```

```python
# Define prompt template
template = """You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use two sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
rag_chain=(
    {"context":retriever, "question":RunnablePassthrough()}

          | prompt
          | llm
          | StrOutputParser()
)
```

```python

res.text
```

```python
questions=["what did the President say about Justic Breyer?",
           "What did the President say abou Intel's CEO?",
           "What did the President say about gun violence?"

           ]
```

```python
groud_truths=[["The president said that Justice Breyer has dedicated his life to serve the country and thanked him for his service."],
              ["The president said that Pat Gelsinger is ready to increase Intel's investment to $100 billion."],
              ["The president asked Congress to pass proven measures to reduce gun violence."]]
```

```python
query =questions[0] #"what did the President say about Justic Breyer?"
```

```python
query
```

```python
rag_chain.invoke(query)
```

```python
answer=[]
context=[]
```

```python
for query in questions:
  answer.append(rag_chain.invoke(query))
  context.append([docs.page_content  for docs in retriever.get_relevant_documents(query)])
```

```python
answer
```

```python
context
```

```python
# To dict
data = {
    "question": questions,
    "answer": answer,
    "contexts": context,
    "ground_truths": groud_truths
}


```

```python
!pip install Dataset
```

```python
from datasets import Dataset
dataset=Dataset.from_dict(data)
```

```python
from ragas import evaluate
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    context_recall,
    context_precision,
)
```

```python
result=evaluate(
    dataset=dataset,
    metrics=[
        context_precision,
        context_recall,
        faithfulness,
        answer_relevancy
    ]
)
```
